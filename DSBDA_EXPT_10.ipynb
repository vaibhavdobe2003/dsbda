```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.datasets import load_iris
import warnings
warnings.filterwarnings("ignore")
```

```python
data=load_iris()
```

```python
data
```

```python
 df= pd.DataFrame()
df[data['feature_names']] = data['data']
df['label'] = data['target']
```
